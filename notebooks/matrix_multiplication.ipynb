```python
import torch

X = torch.Tensor([[2.0, 4.0, 1.0], 
                  [5.0, 6.0, 4.0]])

Y = torch.Tensor([[4.0, 9.0],
                  [2.0, 1.0],
                  [1.0, 2.0]])

tensor_matmul = (X @ Y).numpy().tolist()

print(tensor_matmul)
```

```python
X = X.numpy().tolist()

Y = Y.numpy().tolist()

def dot_product(x, y):
    z = 0
    for a, b in zip(x, y):
        z += a * b
    return z

def transpose(X):
    rows, cols = len(X), len(X[0])
    X_T = [[0 for _ in range(rows)] for _ in range(cols)]
    for row in range(rows):
        for col in range(cols):
            X_T[col][row] = X[row][col]
    return X_T
            

def matrix_multiply(X, Y):
    rows, cols = len(X), len(Y[0])
    Z = [[0 for _ in range(cols)] for _ in range(rows)]
    Y_T = transpose(Y)
    for row in range(rows):
        for col in range(cols):
            Z[row][col] = dot_product(X[row], Y_T[col])
    return Z


manual_matmul = matrix_multiply(X, Y)

print(manual_matmul)
```
